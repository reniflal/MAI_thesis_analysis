```python
import os
import json
import matplotlib.pyplot as plt
import numpy as np
# Specify the file name
file_name = "arne_real_03_05_tt1"
# file_name = "arne_imag_03_05_tt3"
# file_name = "ly_real_01"
# file_name = "ly_real_02"
# file_name = "ly_imag_01"
# file_name = "ece_real_02"
# file_name = "qiang_real_01"
# file_name = "qiang_imag_01"

# file_name = "ly_imag_03"
# Load the dictionary from the file
full_file_path = 'graph/'+file_name+'_eye.json'
with open(full_file_path, 'r') as file:
    eye_data_for_graph = json.load(file)

# Specify the file name
# file_name = 'eeg_data_for_graph.json'

# Load the dictionary from the file
full_file_path = 'graph/'+file_name+'_eeg.json'
with open(full_file_path, 'r') as file:
    eeg_data_for_graph = json.load(file)
    
TP_eye = np.array(eye_data_for_graph["sum_true_pos"])
TP_eeg = np.array(eeg_data_for_graph["sum_true_pos"])
FP_eye = np.array(eye_data_for_graph["sum_false_pos"])
FP_eeg = np.array(eeg_data_for_graph["sum_false_pos"])
TN_eye = np.array(eye_data_for_graph["sum_true_neg"])
TN_eeg = np.array(eeg_data_for_graph["sum_true_neg"])
FN_eye = np.array(eye_data_for_graph["sum_false_neg"])
FN_eeg = np.array(eeg_data_for_graph["sum_false_neg"])

F1_eye = (TP_eye+TP_eye)/(TP_eye+TP_eye+FP_eye+FN_eye)
F1_eeg = (TP_eeg+TP_eeg)/(TP_eeg+TP_eeg+FP_eeg+FN_eeg)
# Precision_eye = 
Accuracy_eye = (TP_eye+TN_eye)/(TP_eye+TN_eye+FP_eye+FN_eye)
Accuracy_eeg = (TP_eeg+TN_eeg)/(TP_eeg+TN_eeg+FP_eeg+FN_eeg)

# Plotting the line chart
plt.figure(figsize=(12, 8))
plt.plot(eeg_data_for_graph["Dwelltime"], eeg_data_for_graph["count_true_positive"], marker='s', linestyle='-', color='r', label='eeg_plus_dwelltime')
plt.plot(eye_data_for_graph["Dwelltime"], eye_data_for_graph["count_true_positive"], marker='o', linestyle='-', color='b', label='simple_dwelltime')

# Adding titles and labels
plt.title('Simple Dwelltime paradigm Vs Dwelltime+BCI paradigm (positive precision)', fontsize=16)
plt.xlabel('Dwelltime', fontsize=14)
plt.ylabel('Percentage', fontsize=14)
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.ylim(0,1.1)
# Create the directory if it does not exist
os.makedirs(os.path.dirname("pics/"+file_name+"/pos_precision_cmpr.png"), exist_ok=True)
# Displaying the chart
plt.savefig("pics/"+file_name+"/pos_precision_cmpr.png")
plt.show()


# Plotting the line chart
plt.figure(figsize=(12, 8))
plt.plot(eeg_data_for_graph["Dwelltime"], eeg_data_for_graph["count_true_negative"], marker='s', linestyle='-', color='r', label='eeg_plus_dwelltime')
plt.plot(eye_data_for_graph["Dwelltime"], eye_data_for_graph["count_true_negative"], marker='o', linestyle='-', color='b', label='simple_dwelltime')

# Adding titles and labels
plt.title('Simple Dwelltime paradigm Vs Dwelltime+EEG paradigm (negative precision)', fontsize=16)
plt.xlabel('Dwelltime', fontsize=14)
plt.ylabel('Percentage', fontsize=14)
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.ylim(0,1.1)
# Displaying the chart
plt.savefig("pics/"+file_name+"/neg_precision_cmpr.png")
plt.show()


# Plotting the line chart
plt.figure(figsize=(12, 8))
plt.plot(eeg_data_for_graph["Dwelltime"], eeg_data_for_graph["count_false_positive"], marker='s', linestyle='-', color='r', label='eeg_plus_dwelltime')
plt.plot(eye_data_for_graph["Dwelltime"], eye_data_for_graph["count_false_positive"], marker='o', linestyle='-', color='b', label='simple_dwelltime')

# Adding titles and labels
plt.title('Simple Dwelltime paradigm Vs Dwelltime+EEG paradigm (sensitivity)', fontsize=16)
plt.xlabel('Dwelltime', fontsize=14)
plt.ylabel('Percentage', fontsize=14)
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.ylim(0,1.1)

# Displaying the chart
plt.savefig("pics/"+file_name+"/sensitivity_cmpr.png")
plt.show()



# Plotting the line chart
plt.figure(figsize=(12, 8))
plt.plot(eeg_data_for_graph["Dwelltime"], eeg_data_for_graph["count_false_negative"], marker='s', linestyle='-', color='r', label='eeg_plus_dwelltime')
plt.plot(eye_data_for_graph["Dwelltime"], eye_data_for_graph["count_false_negative"], marker='o', linestyle='-', color='b', label='simple_dwelltime')

# Adding titles and labels
plt.title('Simple Dwelltime paradigm Vs Dwelltime+EEG paradigm (specificity)', fontsize=16)
plt.xlabel('Dwelltime', fontsize=14)
plt.ylabel('Percentage', fontsize=14)
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.ylim(0,1.1)
# Displaying the chart
plt.savefig("pics/"+file_name+"/specificity_cmpr.png")
plt.show()
```

```python
# Plotting the line chart
plt.figure(figsize=(12, 8))
plt.plot(eeg_data_for_graph["Dwelltime"], F1_eeg, marker='s', linestyle='-', color='r', label='eeg_plus_dwelltime')
plt.plot(eye_data_for_graph["Dwelltime"], F1_eye, marker='o', linestyle='-', color='b', label='simple_dwelltime')

# Adding titles and labels
plt.title('Simple Dwelltime paradigm Vs Dwelltime+EEG paradigm (F1 score)', fontsize=16)
plt.xlabel('Dwelltime', fontsize=14)
plt.ylabel('Percentage', fontsize=14)
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.ylim(0,1.1)
# Displaying the chart
plt.savefig("pics/"+file_name+"/f1_cmpr.png")
plt.show()


# Plotting the line chart
plt.figure(figsize=(12, 8))
plt.plot(eeg_data_for_graph["Dwelltime"], Accuracy_eeg, marker='s', linestyle='-', color='r', label='eeg_plus_dwelltime')
plt.plot(eye_data_for_graph["Dwelltime"], Accuracy_eye, marker='o', linestyle='-', color='b', label='simple_dwelltime')

# Adding titles and labels
plt.title('Simple Dwelltime paradigm Vs Dwelltime+EEG paradigm (Accuracy of Prediction)', fontsize=16)
plt.xlabel('Dwelltime', fontsize=14)
plt.ylabel('Percentage', fontsize=14)
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.ylim(0,1.1)
# Displaying the chart
plt.savefig("pics/"+file_name+"/accuracy_cmpr.png")
plt.show()
```
